```python
# imputing missing values

# numerical feature → use median
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].median())

# numerical feature → use median
dataset['MonthlyIncome'] = dataset['MonthlyIncome'].fillna(dataset['MonthlyIncome'].median())

# categorical feature → use most frequent value (mode)
dataset['TypeofContact'] = dataset['TypeofContact'].fillna(dataset['TypeofContact'].mode()[0])

# numerical feature → use median
dataset['DurationOfPitch'] = dataset['DurationOfPitch'].fillna(dataset['DurationOfPitch'].median())

# discrete feature → use mode
dataset['NumberOfFollowups'] = dataset['NumberOfFollowups'].fillna(dataset['NumberOfFollowups'].mode()[0])

# discrete feature → use mode
dataset['PreferredPropertyStar'] = dataset['PreferredPropertyStar'].fillna(
    dataset['PreferredPropertyStar'].mode()[0]
)

# discrete feature → use mode
dataset['NumberOfTrips'] = dataset['NumberOfTrips'].fillna(dataset['NumberOfTrips'].mode()[0])

# discrete feature → use mode
dataset['NumberOfChildrenVisiting'] = dataset['NumberOfChildrenVisiting'].fillna(
    dataset['NumberOfChildrenVisiting'].mode()[0]
)
```

```python
dataset.isnull().sum()
```

```python
# drop customerId
dataset.drop('CustomerID',axis=1,inplace=True)
```

```python
dataset.head()
```

### Feature Engineering

```python
# create new column for feature NumberOfPersonVisiting and NumberOfChildrenVisiting
dataset['TotalNumberOfVisitors'] = dataset['NumberOfPersonVisiting'] + dataset['NumberOfChildrenVisiting']
dataset.drop(['NumberOfPersonVisiting','NumberOfChildrenVisiting'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
# get all categorical columns
categorical_columns = dataset.select_dtypes(include=['object']).columns
categorical_columns
```

```python
# get all numerical columns
numerical_columns = dataset.select_dtypes(include=['int64','float64']).columns
numerical_columns
```

```python
# get all discrete columns
discrete_columns = [features for features in numerical_columns if len(dataset[features].unique())<25]
discrete_columns
```

```python
# get all continuous columns
continuous_columns = [features for features in numerical_columns if features not in discrete_columns]
continuous_columns
```

```python
# save the cleaned dataset
dataset.to_csv('cleaned_holiday_package_data.csv',index=False)
```
